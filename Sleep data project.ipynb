```python
# how many sleep disorder are present in the dataset
disorder_column = "Sleep Disorder"  

# Find unique sleep disorders
unique_disorders = df[disorder_column].unique()

# Count the number of occurrences of each disorder
disorder_counts = df[disorder_column].value_counts()

# Print results
print(f"Unique Sleep Disorders Present: {len(unique_disorders)-1}\n")
print("Sleep Disorder Counts:")
print(disorder_counts)
```

```python
#print(df.columns)
print("INSOMNIA DETAILS:\n")
# Find the correct disorder column
correct_column_name = "Sleep Disorder"  

# Total number of people
total_people = len(df)

# Number of people with Insomnia
insomnia_people = df[df[correct_column_name] == "Insomnia"]

# Percentage of Insomnia cases
insomnia_percentage = (len(insomnia_people) / total_people) * 100

# Average sleep duration for Insomnia cases
avg_sleep_duration = insomnia_people["Sleep Duration"].mean()

avg_heart_rate= insomnia_people["Heart Rate"].mean()

avg_physical= insomnia_people["Physical Activity Level"].mean()

avg_stress= insomnia_people["Stress Level"].mean()

#blood pressure
bp_split = insomnia_people["Blood Pressure"].str.split("/", expand=True)

# Convert to numeric values
insomnia_people["Systolic BP"] = pd.to_numeric(bp_split[0])
insomnia_people["Diastolic BP"] = pd.to_numeric(bp_split[1])

# Calculate average Systolic and Diastolic BP
avg_systolic_bp = insomnia_people["Systolic BP"].mean()
avg_diastolic_bp = insomnia_people["Diastolic BP"].mean()
avg_daily= insomnia_people["Daily Steps"].mean()
avg_qlsleep=insomnia_people["Quality of Sleep"].mean()
# Print results
print(f"Percentage of people with Insomnia: {insomnia_percentage:.2f}%\n")
print(f"Average Sleep Duration for Insomnia cases: {avg_sleep_duration:.2f} hrs\n")
print(f"Average Heart Rate for Insomnia cases: {avg_heart_rate:.2f} bpm\n")
print(f"Average Physical Activity Level for Insomnia cases: {avg_physical:.2f}\n")
print(f"Average Stress Level for Insomnia cases: {avg_stress:.2f}\n")
print(f"Average Blood Pressure for Insomnia cases: {avg_systolic_bp:.2f}/{avg_diastolic_bp:.2f}\n")
print(f"Average Daily steps for Insomnia cases: {avg_daily:.2f}\n")
print(f"Average Quality of Sleep for Insomnia cases: {avg_qlsleep:.2f} hrs\n")
```

```python
# male and Female percentage
gender_column = "Gender"      
disorder_column = "Sleep Disorder"  

# Filter only Insomnia patients
insomnia_people = df[df[disorder_column] == "Insomnia"]

# Count male and female insomnia patients
insomnia_gender_count = insomnia_people[gender_column].value_counts()

# Calculate percentage
total_insomnia = len(insomnia_people)
male_percentage = (insomnia_gender_count.get("Male", 0) / total_insomnia) * 100
female_percentage = (insomnia_gender_count.get("Female", 0) / total_insomnia) * 100

# Print results
print(f"Total Insomnia Patients: {total_insomnia}")
print(f"Male Insomnia Patients: {insomnia_gender_count.get('Male', 0)} ({male_percentage:.2f}%)")
print(f"Female Insomnia Patients: {insomnia_gender_count.get('Female', 0)} ({female_percentage:.2f}%)")
```

```python
occupation_column = "Occupation"  
disorder_column = "Sleep Disorder"     
# Filter only Insomnia patients
insomnia_people = df[df[disorder_column] == "Insomnia"]

# Count occupation-wise Insomnia cases
insomnia_occupation_count = insomnia_people[occupation_column].value_counts()

# Calculate percentage
total_insomnia = len(insomnia_people)
insomnia_occupation_percentage = (insomnia_occupation_count / total_insomnia) * 100

# Print results
print("Insomnia Patients by Occupation:")
print(insomnia_occupation_count)
print("\nPercentage of Insomnia Patients by Occupation:")
print(insomnia_occupation_percentage)
```

```python
bmi_column = "BMI Category"  
disorder_column = "Sleep Disorder" 

# Filter only Insomnia patients
insomnia_people = df[df[disorder_column] == "Insomnia"]

# Count BMI Category-wise Insomnia cases
insomnia_bmi_count = insomnia_people[bmi_column].value_counts()

# Calculate percentage of Insomnia cases in each BMI Category
total_insomnia = len(insomnia_people)
insomnia_bmi_percentage = (insomnia_bmi_count / total_insomnia) * 100

# Print results
print("Insomnia Cases by BMI Category:")
print(insomnia_bmi_count)
print("\nPercentage of Insomnia Cases by BMI Category:\n")
print(insomnia_bmi_percentage)
```
